#### Грузим данные:

```python
with open('text.txt', 'r', encoding = 'utf-8') as src:
    whole_text = src.read()
    whole_text = re.sub('\n', ' ', whole_text)
    whole_text = re.sub('  ', ' ', whole_text)
with open('text_alligned.txt', 'r', encoding= 'utf-8') as src_gold:
    gold = src_gold.read()
    gold = gold.split('\n')
with open('to_compare.txt', 'r', encoding = 'utf-8') as src_comp:
    comparison = src_comp.read()
```

Получившиеся переменные:
* whole_text - текст всего произведения одной строкой.
* gold - выборка из 50 первых предложений, разбитых вручную разделителями строки
* comparison - 50 первых предложений, разделенные пробелами

#### Делаем функцию, разбивающую текст регулярками:

```python
def my_tokenizer(t):
    new_t = re.sub(r'([\.\?!]) ([А-ЯЁ])', r'\1\n\2', t)
    new_t = re.split(r'\n', new_t)
    return new_t
```

```python
my_split = my_tokenizer(comparison)
```

```python
my_split
```

```python
print('Gold length - ', len(gold))
print('My split - ', len(my_split))
overlap = len(set(gold) & set(my_split)) / len(set(gold) | set(my_split))
print('percentage of overlaps - '+str(overlap))
```

```python
tp = 0
fp = 0
fn = 0

for sent in gold:
    if len(my_tokenizer(sent)) == 1:
        tp += 1
    else:
        fp += 1

for i in range(len(gold)-1):
    sent1, sent2 = gold[i], gold[i+1]
    sent = ' '.join([sent1, sent2])
    if len(my_tokenizer(sent)) == 2:
        tp += 1
    else:
        fn += 1

precision = (tp/(tp+fp))
recall = (tp/(tp+fn))
f1 = 2*(precision*recall)/(precision+recall)
print('Precision - ', precision)
print('Recall - ', recall)
print('F1 - ', f1)
```

Вывод: главная проблема - неучтенные предложения, стоящие на границе прямой и косвенной речи. Подправим код:

```python
def my_tokenizer_2(t):
    new_t = re.sub(r'([\.\?!]) ([А-ЯЁ])', r'\1\n\2', t)
    new_t = re.sub(r'([\.\?!:])(»| —) ([А-ЯЁ])', r'\1\n\3', new_t)
    #new_t = re.sub(r'([\.\?!])» ([А-ЯЁ])', r'', t)
    new_t = re.split(r'\n', new_t)
    return new_t
```

```python
my_split_2 = my_tokenizer_2(comparison)
```

#### PunktSentenceTokenizer 

```python
from pprint import pprint
from nltk.tokenize.punkt import PunktSentenceTokenizer, PunktTrainer
```

```python
trainer = PunktTrainer()
trainer.INCLUDE_ALL_COLLOCS = True
trainer.train(whole_text)
tokenizer = PunktSentenceTokenizer(trainer.get_params())
```

```python
tokenizer_split = tokenizer.tokenize(comparison)
```

```python
print('Gold length - ', len(gold))
print('Tokenizer\'s split - ', len(tokenizer_split))
overlap_3 = len(set(gold) & set(tokenizer_split)) / len(set(gold) | set(tokenizer_split))
print('percentage of overlaps - '+str(overlap_3))
```

```python
tp = 0
fp = 0
fn = 0

for sent in gold:
    if len(tokenizer.tokenize(sent)) == 1:
        tp += 1
    else:
        fp += 1

for i in range(len(gold)-1):
    sent1, sent2 = gold[i], gold[i+1]
    sent = ' '.join([sent1, sent2])
    if len(tokenizer.tokenize(sent)) == 2:
        tp += 1
    else:
        fn += 1

precision = (tp/(tp+fp))
recall = (tp/(tp+fn))
f1 = 2*(precision*recall)/(precision+recall)
print('Precision - ', precision)
print('Recall - ', recall)
print('F1 - ', f1)
```
